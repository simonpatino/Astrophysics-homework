```python
altitude = np.linspace(0 ,100, 1000)



coesa76_geom = coesa76(altitude) # geometric altitudes by default

plt.plot(coesa76_geom.T,altitude)

plt.xlabel("Temperatura(K)")
plt.ylabel("Altura(km)")

plt.grid()
```

```python
plt.plot(coesa76_geom.rho,altitude)

plt.xlabel("$\\rho$ $(kg/m^3)$ ")
plt.ylabel("Altura(km)")

plt.grid()
```

```python
plt.plot(coesa76_geom.P/1000,altitude)

plt.xlabel("Presion ($KPa$) ")
plt.ylabel("Altura(km)")
plt.grid()
```

```python
import numpy as np

def atm_mars(h_list):
    temperatures = []
    pressures = []
    densities = []

    for h in h_list:
        if h > 7:
            t = -23.4 - 0.00222 * h
            p = 0.699 * np.exp(-0.00009 * h)
        else:
            t = -31 - 0.000998 * h
            p = 0.699 * np.exp(-0.00009 * h)

        rho = p / (0.1921 * (t + 273.1))

        temperatures.append(t)
        pressures.append(p)
        densities.append(rho)

    return temperatures, pressures, densities

# Example usage
#altitudes = [5000, 8000, 10000]
#temperatures, pressures, densities = atm_mars(altitudes)
#for h, t, p, rho in zip(altitudes, temperatures, pressures, densities):
   # print(f"At altitude {h} meters on Mars:")
  #  print(f"Temperature: {t} K")
 #   print(f"Pressure: {p} Pa")
#    print(f"Density: {rho} kg/m^3")
```
