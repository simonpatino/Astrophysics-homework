#*Perfil de temperatura para diferentes masas solares*

```python
M = 1 # SOLAR MASS

M2 = 2

M1_2 = 0.5

tM = lambda r: 280*(r)**(-1/2)*(M**4)**(1/4)
tM2 = lambda r: 280*(r)**(-1/2)*(M2**4)**(1/4)
tM1_2 =  lambda r: 280*(r)**(-1/2)*(M1_2**4)**(1/4)

r =  np.linspace(0.1,100,1000)


tM = tM(r)

tM2= tM2(r)

tM1_2= tM1_2(r)


plt.plot(r, tM , label = " 1 Solar Mass")

plt.plot(r, tM2, label = " 2 Solar Mass")

plt.plot(r, tM1_2, label = " 0.5 Solar Mass")

plt.legend()

plt.xlabel("r(AU)")

plt.ylabel("T(k)")

plt.show()
```

#*Desaparición del disco*

```python
fg = lambda t: np.exp(-t/(1e7))

sigma_g_0= lambda r:  75*fg(0)*(r/(10))**(-1/2)

sigma_g_1= lambda r:  75*fg(1e5)*(r/(10))**(-1/2)

sigma_g_2= lambda r:  75*fg(1e6)*(r/(10))**(-1/2)

r =  np.linspace(0.01,3000,1000)

sigma0 = sigma_g_0(r)

sigma1 = sigma_g_1(r)

sigma2 = sigma_g_2(r)

plt.plot(r, sigma0 , label="$t=0  \ years$")

plt.plot(r, sigma1, label="$t=10^5 \ years$")

plt.plot(r, sigma2, label="$t=10^6 \ years$")

plt.xscale("log")

plt.yscale("log")

plt.xlabel("r[au]")

plt.ylabel("$\Sigma_g $")

plt.legend()
```

```python
ro =  np.linspace(0.01,2.7,1000)

fd = 1

sigmad_pre = lambda r: 0.32*fd*(r/10)**(-1/2)

sigmad_pres =  sigmad_pre(ro)

rn =  np.linspace(2.7,200,1000)

sigmad_post = lambda r: 0.32*fd*(r/10)**(-1/2)*4.2

sigmad_posts = sigmad_post(rn)

sigma =  np.concatenate( (sigmad_pres, sigmad_posts), axis = 0 )

#plt.plot(ro, sigmad_pres , label="$t=0  \ years$")

#plt.plot(rn, sigmad_posts , label="$t=0  \ years$")


plt.plot( np.concatenate((ro,rn), axis=0) , sigma , label="$t=0  \ years$")




plt.xscale("log")

plt.yscale("log")

plt.xlabel("r[au]")

plt.ylabel("$\Sigma_d $")

plt.legend()

```
