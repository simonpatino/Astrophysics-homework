```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
M = 1 # SOLAR MASS

M2 = 2

M1_2 = 0.5

tM = lambda r: 280*(r)**(-1/2)*(M**4)**(1/4)
tM2 = lambda r: 280*(r)**(-1/2)*(M2**4)**(1/4)
tM1_2 =  lambda r: 280*(r)**(-1/2)*(M1_2**4)**(1/4)

r =  np.linspace(0.1,100,1000)


tM = tM(r)

tM2= tM2(r)

tM1_2= tM1_2(r)


plt.plot(r, tM , label = " 1 Solar Mass")

plt.plot(r, tM2, label = " 2 Solar Mass")

plt.plot(r, tM1_2, label = " 0.5 Solar Mass")

plt.legend()


plt.xlabel("r(AU)")

plt.ylabel("T(k)")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the size of the grid
n = 100
x = np.linspace(-1, 1, n)
y = np.linspace(-1, 1, n)
X, Y = np.meshgrid(x, y)

# Define the point and its temperature
point = (0, 0)
temperature_at_point = 100

# Calculate the temperature at each point on the grid
distance = np.sqrt((X - point[0])**2 + (Y - point[1])**2)
temperature = temperature_at_point / (1 + distance)

# Plot the filled contour plot
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, temperature, cmap='coolwarm')
plt.colorbar(label='Temperature')
plt.scatter(*point, color='red', label='Point')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Temperature Change around a Point')
plt.axis('equal')  # Equal aspect ratio
plt.grid(True)
plt.show()
```
